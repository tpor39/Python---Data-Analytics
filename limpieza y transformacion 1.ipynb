<p style="color: green; font-weight: bold; font-size: 24px;">PARCEO, para transpormar string a numerico</p>

<p style="color: red; font-weight: bold; font-size: 17px;">Le pedimos al chat gpt lo siguiente:.</p>

Necesito parcear la columna salary de un data set que contiene salario en forma de string. pero necesito crear una columna numerica con el valor numerico del salario en cuestion. este es un ejemplo del valor de la string de una fila en la columna salary

$100,000.00 
       - 
       $120,000.00
como puedo obtener un valor numerico de 100000?

```python
import pandas as pd
import re

# FunciÃ³n para extraer el valor numÃ©rico del salario mÃ­nimo
def extract_min_salary(salary_str):
    # Usar una expresiÃ³n regular para encontrar el primer valor monetario
    match = re.search(r'\$([\d,]+\.?\d*)', salary_str)
    if match:
        min_salary = match.group(1)
        # Eliminar comas
        min_salary = min_salary.replace(',', '')
        return float(min_salary)
    return None

# Aplicar la funciÃ³n para crear una nueva columna con el salario mÃ­nimo en el DataFrame df_datos_sin_nulos
df_sin_nulos['min_salary'] = df_sin_nulos['salary'].apply(extract_min_salary)

# Mostrar el DataFrame resultante
df_sin_nulos[['salary', 'min_salary']]
```

```python
df_sin_nulos
```

<p style="color: BLUE; font-weight: bold; font-size: 24px;">Creamos un filtro con el sig codigo.</p>

```python
#nos quedamos con los salarios mayores a 20000
df_sin_nulos = df_sin_nulos[df_sin_nulos['min_salary']>= 20000]
```

<p style="color: green; font-weight: bold; font-size: 24px;">recien ahora podemos rellenar los daton no nulos con valores promedios ya que transformamos de string a numerico en la columna min_salary.</p>

<p style="color: red; font-weight: bold; font-size: 24px;">lo hacemos con la funcion fillna.</p>

```python
df_con_relleno = df_sin_nulos.fillna(df_sin_nulos['min_salary'].mean())

df_con_relleno['min_salary'].describe()
```

```python
df_con_relleno
```

<p style="color: green; font-weight: bold; font-size: 24px;">eliminamos columna que no sirve.</p>

```python
df_con_relleno.drop('salary', axis=1, inplace=True)
```

```python
df_con_relleno
```

<p style="color: BLUE; font-weight: bold; font-size: 24px;">AGRUPAR Y AGREGAR DATOS.</p>

```python
# Agrupar por 'title' y calcular la media, mediana y conteo de 'min_salary'
df_grouped = df_con_relleno.groupby('title')['min_salary'].agg(['mean', 'median', 'count' ]).reset_index()

# Renombrar las columnas resultantes
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'job_count']

# Mostrar el DataFrame agrupado
df_grouped
```
