<div class="stylish-text">GROUP BY</div>

CREAR UNA TABLA QUE NOS MUESTRE EL TOP 10 DE VALOR PROMEDIO DE MERCADO DE JUGADORES AGRUPANDO POR CLUB ACTUAL Y ORDENADO DE MAYO A MENOR PROMEDIO

```python
#Configurar pandas para mostrar numeros en formato completo
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
q= """

SELECT
current_club_name as nombre_del_club,

avg(highest_market_value_in_eur) as promedio_valor

FROM jugadores

GROUP BY 1 

ORDER BY promedio_valor DESC

LIMIT 11

"""

top_10_clubes = pysqldf(q)
```

<div class="stylish-text">HAVING</div>

APLICAMOS UN HAVING PARA QUE TOME SOLO LOS VALORES DEL PROMEDIO_VALOR <1500000

SE APLICA HAVING PORQUE LA CONDICION LA ESTAMOS APLICANDO SOBRE EL VALOR FILTRADO EN EL GROUPBY. SI NO HUBIESE NINGUN VALOR FILTRADO UTILIZARIAMOS LA COMANDA WJERE

```python
q= """

SELECT
current_club_name as nombre_del_club,

avg(highest_market_value_in_eur) as promedio_valor

FROM jugadores

GROUP BY 1 

HAVING promedio_valor < 15000000

ORDER BY promedio_valor DESC

LIMIT 11

"""

top_10_clubes = pysqldf(q)
```

<div class="stylish-text">LEFT JOIN</div>

SUPONGAMOS QUE PARTIMOS DE LA TABLA DE DE LOS TOP 10 SEGUN VALUACION PROMEDIO DE JUGADORES 

```python
top_10_clubes
```

AHORA LO QUE QUEREMOS SABER ES COMO SE LLAMA CADA UNO DE LOS ESTADIOS DE ESTOS CLUBES, PARA ESO UNIMOS ESTA TABLA CON LA TABLA CLUBES

```python
clubes
```

ON a.nombre_del_club = b.name   es la key que aparece en las 2 columnas

```python
q = """ 
SELECT a.*,

 b.stadium_name

FROM top_10_clubes a 

LEFT JOIN  clubes b 

ON a.nombre_del_club = b.name

"""

pysqldf(q)
```
