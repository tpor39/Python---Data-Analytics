<div class="stylish-text">Limpieza de datos utilizando el IQR 
      Como orga3!!!
</div>

Este proceso esta relacionado con la identificacion de outliers en un conjunto de datos. Aqui la explicacion detallada paso a paso. 

1. Calculo de los cuartiles (Q1,Q3):
    * Q1 representa el primer cuartil que es el valor por debajo del cual se encuentra el 25% de los datos
    * Q3 representa el tercer cuartil que es el valor por debajo del cual se encuentra el 75% de los datos
    * Estos cuartiles dividen ek conjunto de datos en cuatro partes iguales
2. Calculo del rango intercuartilico (IQR)
    * El rango intercuartilico es la diferencia entre el tercer cuartil(Q3) y el primer cuartil(Q1)
    * Matematicamente IQR=Q3-Q1
3. Definicion de límites para Outliers:
    * Se establecen límites superior e inferior para identificar valores atipicos de los datos.
    * El limite inferior se calcula restando 1.5 veces el IQR a Q1: Lower_bound = Q1-1.5*IQR
    * El limite superior se calcula sumando 1.5 veces el IQR a Q3: Upper_bound = Q3+1.5*IQR
4. Identificacion de Outliers:
    * Cualquier valor que este fuera de Lowerbound / Upperbound es un outlier

La idea de identicar estas cuasas especiales de variacion es identificarlos y darles un analisis independiente a cada caso.     

```python
#calcular rango intercuartilico (IQR)
Q1= df['precio_en_euros'].quantile(0.25)
Q3 = df['precio_en_euros'].quantile(0.75)
IQR = Q3-Q1

#definir los limites para considerar outliers

lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

#filtrar el dataframe para obtener solo los valores que se encuentren dentro de los límites.
df_filtered = df[(df['precio_en_euros'] >= lower_bound) & (df['precio_en_euros'] <= upper_bound)]
```

```python
df_filtered
```

```python
#ajusta la escala del eje y muestra numeros completos
plt.ticklabel_format(style='plain', axis='y')

#boxplot con matplotlib
plt.boxplot(df_filtered['precio_en_euros'])
plt.title('boxplot de precio_en_euros')
plt.show()

#histograma con matplotlib

plt.hist(df_filtered['precio_en_euros'], bins=100)
plt.title('histograma en euros')
plt.show()
```

<div class="stylish-text">Correlación</div>

1. La correlacion entre 2 variables es una medida estadistica que describe la relacion estadistica y la dependencia entre ellas. En otras palabras la correlacion indica como cambia una variable en relacion con la otra.
    + La correlacion mide la relacion lineal entre dos variables. El coeficiente de correlacion varia entre [-1;1]
        * Si es 1, significa correlacion positiva perfecta
        * Si es -1, indica una correlacion negativa perfecta
        * Si es 0 no hay correlacion lineal

```python
#creamos un nuevo dataframe solo con variables numericas para poder calcular la correlacion

df_nuevo = df[['precio_en_euros', 'club_actual_id']].copy()
```

```python
#calculo la matriz de correlacion
correlation_matrix = df_nuevo.corr()

#visualizacion del heatmap con Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('matriz de correlacion')
plt.show()
```
